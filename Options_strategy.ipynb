```python
import numpy as np
import matplotlib.pyplot as plt 
```

**Long Call**

```python
prenium=5
S0=95
K=102
St=np.linspace(0,150,150)
pnl=np.maximum(St-K,0)-prenium


plt.plot(St, pnl)
plt.axvline(K, color='green', linestyle='--', linewidth=0.5, label='Strike')
plt.axvline(S0, color='orange', linestyle='--', linewidth=0.5, label='Intial Stock Price')
plt.xlabel('price')
plt.ylabel('P&L')
plt.title('Profit and Loss of a Long Call')
plt.grid()
plt.show()
```

**Short Call**

```python
prenium=5
S0=95
K=102
St=np.linspace(0,150,150)
pnl=-np.maximum(St-K,0)+prenium


plt.plot(St, pnl)
plt.axvline(K, color='green', linestyle='--', linewidth=0.5, label='Strike')
plt.axvline(S0, color='orange', linestyle='--', linewidth=0.5, label='Intial Stock Price')
plt.xlabel('price')
plt.ylabel('P&L')
plt.title('Profit and Loss of a Short Call')
plt.grid()
plt.show()
```

**Long Put**

```python
St=np.linspace(0,150,150)
S0=95
K=102
prenium=5

pnl=np.maximum(K-St,0)-prenium

plt.plot(St,pnl)
plt.axvline(K,color='green', linestyle='--',linewidth=0.5,label='Strike')
plt.axvline(S0, color='orange', linestyle='--', linewidth=0.5, label='Intial Stock Price')
plt.xlabel('underlying price')
plt.ylabel('P&L')
plt.title('Profit and Loss of Long Put')
plt.grid()
plt.show()
```
