```python
import numpy as np
import matplotlib.pyplot as plt 
```

First let's start by visualizing calls and puts when shorting and when long, then we will combine this to create different basic strategies.

### Long Call

Payoff:  
$$ P(S_T) = \max(S_T - K, 0) $$

P&L:  
$$ P\&L(S_T) = \max(S_T - K, 0) - C $$

Maximum loss: $-C$  
Maximum gain: unlimited 

```python
prenium=5
S0=95
K=102
St=np.linspace(0,150,150)
pnl=np.maximum(St-K,0)-prenium


plt.plot(St, pnl, label='P&L')
plt.axvline(K, color='green', linestyle='--', linewidth=0.5, label='Strike')
plt.axvline(S0, color='orange', linestyle='--', linewidth=0.5, label='Intial Stock Price')
plt.xlabel('Underlying price')
plt.ylabel('P&L')
plt.title('Profit and Loss of a Long Call')
plt.legend()
plt.grid()
plt.show()
```

<br><br>
### Short Call

Payoff:  
$$ P(S_T) = -\max(S_T - K, 0) $$

P&L:  
$$ P\&L(S_T) = -\max(S_T - K, 0) + C $$

Maximum loss: unlimited  
Maximum gain: $+C$ (premium received) 

```python
prenium=5
S0=95
K=102
St=np.linspace(0,150,150)
pnl=-np.maximum(St-K,0)+prenium


plt.plot(St,pnl, label='P&L')
plt.axvline(K, color='green', linestyle='--', linewidth=0.5, label='Strike')
plt.axvline(S0, color='orange', linestyle='--', linewidth=0.5, label='Intial Stock Price')
plt.xlabel('Underlying price')
plt.ylabel('P&L')
plt.title('Profit and Loss of a Short Call')
plt.legend()
plt.grid()
plt.show()
```

<br><br>
### Long Put

Payoff:  
$$ P(S_T) = \max(K - S_T, 0) $$

P&L:  
$$ P\&L(S_T) = \max(K - S_T, 0) - P $$

Maximum loss: $-P$  
Maximum gain: $K - P$ (if $S_T \to 0$)  


```python
St=np.linspace(0,150,150)
S0=95
K=102
prenium=5

pnl=np.maximum(K-St,0)-prenium

plt.plot(St,pnl, label='P&L')
plt.axvline(K,color='green', linestyle='--',linewidth=0.5,label='Strike')
plt.axvline(S0, color='orange', linestyle='--', linewidth=0.5, label='Intial Stock Price')
plt.xlabel('Underlying price')
plt.ylabel('P&L')
plt.title('Profit and Loss of Long Put')
plt.legend()
plt.grid()
plt.show()
```

<br><br>
### Short Put

Payoff:  
$$ P(S_T) = -\max(K - S_T, 0) $$

P&L:  
$$ P\&L(S_T) = -\max(K - S_T, 0) + P $$

Maximum loss: $-(K - P)$ (if $S_T \to 0$)  
Maximum gain: $+P$ (premium received)  

```python
St=np.linspace(0,150,150)
S0=95
K=102
prenium=5

pnl=-np.maximum(K-St,0)+prenium

plt.plot(St,pnl, label='P&L')
plt.axvline(K,color='green', linestyle='--',linewidth=0.5,label='Strike')
plt.axvline(S0, color='orange', linestyle='--', linewidth=0.5, label='Intial Stock Price')
plt.xlabel('underlying price')
plt.ylabel('P&L')
plt.title('Profit and Loss of Short Put')
plt.legend()
plt.grid()
plt.show()
```

**Now that we have looked at the basic objects, we will combine them to develop different strategies**

<br><br><br><br>

*We will then look only at the long position, the short position being the reverse graph*
<br>

<br><br>
### Straddle (long call + long put, same strike $K$)

Payoff:  
$$ P(S_T) = \max(S_T - K, 0) + \max(K - S_T, 0) $$

P&L:  
$$ P\&L(S_T) = P(S_T) - (C + P) $$

Maximum loss: $-(C+P)$  
Maximum gain: unlimited if $S_T \to \infty$, or $K - (C+P)$ if $S_T \to 0$  

```python
S0=95 
K=102
St=np.linspace(52,152,100)
prenium=5

payoff_call=np.maximum(St-K,0)
payoff_put=np.maximum(K-St,0)

pnl=payoff_call+payoff_put-prenium

plt.plot(St,pnl, label='P&L')
plt.axvline(K,color='green', linestyle='--',linewidth=0.5,label='Strike')
plt.axvline(S0, color='orange', linestyle='--', linewidth=0.5, label='Intial Stock Price')
plt.xlabel('Underlying price')
plt.ylabel('P&L')
plt.title('Profit and Loss of Straddle')
plt.legend()
plt.grid()
plt.show()
```

<br><br>
### Strangle (long call $K_2$ + long put $K_1$, with $K_1 < K_2$)

Payoff:  
$$ P(S_T) = \max(S_T - K_2, 0) + \max(K_1 - S_T, 0) $$

P&L:  
$$ P\&L(S_T) = P(S_T) - (C + P) $$

Maximum loss: $-(C+P)$  
Maximum gain: unlimited if $S_T \to \infty$, or $K_1 - (C+P)$ if $S_T \to 0$  


```python
S0=90 
K1=100
K2=120
St=np.linspace(70,150,80)
prenium=2

payoff_call=np.maximum(St-K2,0)
payoff_put=np.maximum(K1-St,0)

pnl=payoff_call+payoff_put-prenium

plt.plot(St,pnl, label='P&L')
plt.axvline(K1,color='green', linestyle='--',linewidth=0.5,label='Strike 1 du put')
plt.axvline(K2,color='green', linestyle='--',linewidth=0.5,label='Strike 2 du call')
plt.axvline(S0, color='orange', linestyle='--', linewidth=0.5, label='Intial Stock Price')
plt.xlabel('Underlying price')
plt.ylabel('P&L')
plt.title('Profit and Loss of Strangle')
plt.legend()
plt.grid()
plt.show()
```

<br><br>
### Butterfly Spread (using calls with $K_1 < K_2 < K_3$)

Payoff:  
$$ P(S_T) = \max(S_T - K_1, 0) - 2 \max(S_T - K_2, 0) + \max(S_T - K_3, 0) $$

P&L:  
$$ P\&L(S_T) = P(S_T) - \text{initial cost} $$

Maximum gain: $K_3 - K_2 - \text{initial cost}$ (at $S_T = K_2$)  
Maximum loss: $\text{initial cost}$ 

```python
#achat call K1, vente de 2 call K2 et achat de call K3 avec K1<K2<K3
S0=90
K1=100
K2=110
K3=120
St=np.linspace(70,170,100)
prenium=5

payoff_call1=np.maximum(St-K1,0)
payoff_call2=-2*np.maximum(St-K2,0)
payoff_call3=np.maximum(St-K3,0)

pnl=payoff_call1+payoff_call2+payoff_call3-prenium

plt.plot(St,pnl, label='P&L')
plt.axvline(K1,color='green', linestyle='--',linewidth=0.5,label='Strike 1')
plt.axvline(K2,color='green', linestyle='--',linewidth=0.5,label='Strike 2')
plt.axvline(K3,color='green', linestyle='--',linewidth=0.5,label='Strike 3')
plt.axvline(S0, color='orange', linestyle='--', linewidth=0.5, label='Intial Stock Price')
plt.xlabel('Underlying price')
plt.ylabel('P&L')
plt.title('Profit and Loss of Butterfly')
plt.legend()
plt.grid()
plt.show()
```

<br><br>
### Bull Spread

long call strike $K_1$, short call strike $K_2 > K_1$  

Payoff:  
$$ P(S_T) = \max(S_T - K_1, 0) - \max(S_T - K_2, 0) $$

P&L:  
$$ P\&L(S_T) = P(S_T) - \text{net cost} $$

Maximum gain: $K_2 - K_1 - \text{net cost}$  
Maximum loss: $\text{net cost}$  

```python
S0=90 
K1=100
K2=120
St=np.linspace(70,150,80)
prenium=5

payoff_call1=np.maximum(St-K1,0)
payoff_call2=-np.maximum(St-K2,0)

pnl=payoff_call1+payoff_call2-prenium

plt.plot(St,pnl, label='P&L')
plt.axvline(K1,color='green', linestyle='--',linewidth=0.5,label='Strike 1')
plt.axvline(K2,color='green', linestyle='--',linewidth=0.5,label='Strike 2')
plt.axvline(S0, color='orange', linestyle='--', linewidth=0.5, label='Intial Stock Price')
plt.xlabel('Underlying price')
plt.ylabel('P&L')
plt.title('Profit and Loss of Bull Spread')
plt.legend()
plt.grid()
plt.show()
```

<br><br>
### Bear Spread

long put strike $K_2$, short put strike $K_1 < K_2$  

Payoff:  
$$ P(S_T) = \max(K_2 - S_T, 0) - \max(K_1 - S_T, 0) $$

P&L:  
$$ P\&L(S_T) = P(S_T) - \text{net cost} $$

Maximum gain: $K_2 - K_1 - \text{net cost}$  
Maximum loss: $\text{net cost}$ 

```python
S0=90 
K1=100
K2=120
St=np.linspace(70,150,80)
prenium=5

payoff_put1=np.maximum(K2-St,0)
payoff_put2=-np.maximum(K1-St,0)

pnl=payoff_put1+payoff_put2-prenium

plt.plot(St,pnl, label='P&L')
plt.axvline(K1,color='green', linestyle='--',linewidth=0.5,label='Strike 1')
plt.axvline(K2,color='green', linestyle='--',linewidth=0.5,label='Strike 2')
plt.axvline(S0, color='orange', linestyle='--', linewidth=0.5, label='Intial Stock Price')
plt.xlabel('Underlying price')
plt.ylabel('P&L')
plt.title('Profit and Loss of Bear Spread')
plt.legend()
plt.grid()
plt.show()
```
